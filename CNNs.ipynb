```
#import thu vien
import cv2 as cv
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```
#tao cac ma tran de chua data
trainX = []
trainY = []
testX = []
testY = []

#link cua cac thu muc chua du lieu
train_links = '/content/drive/My Drive/CNN/SAM/Datasets/samtrain/'
test_links = '/content/drive/My Drive/CNN/SAM/Datasets/samtest/'

#load data
def load_data(links):
    X = []
    Y = []

    path_samhanquoc = links + 'samhanquoc//'
    list_link_1 = os.listdir(path_samhanquoc)
    path_samngoclinh = links + 'samngoclinh//'
    list_link_2 = os.listdir(path_samngoclinh)

    for item in list_link_1:
        path = path_samhanquoc + item
        img = cv.imread(path,1)
        img = cv.resize(img, (150,150))
        X.append(img)
        Y.append([1,0])

    for item in list_link_2:
        path = path_samngoclinh + item
        img = cv.imread(path,1)
        img = cv.resize(img, (150,150))
        X.append(img)
        Y.append([0,1])
    X = np.asarray(X)
    Y = np.asarray(Y)
    
    return X , Y

trainX , trainY = load_data(train_links)
testX , testY = load_data(test_links)

#tro du lieu
trainX, trainY = shuffle(trainX, trainY)

tainX = trainX.astype('float') / 255.0
```

```
print(trainX.shape)
```

```
#import cac thu vien can thiet cho model
from keras.models import Sequential
from keras.layers import Conv2D,Activation,BatchNormalization,MaxPool2D,Flatten,Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import SGD
```

```
w = 150
h = 150
classer = 2

shape = (w,h,3)
```

```
#tao mot model
model = Sequential()
```

```
model.add(Conv2D(16,(3,3),padding='valid',input_shape = shape))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Conv2D(16,(3,3),padding='valid'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))
```

```
model.add(Conv2D(32,(3,3),padding='valid'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Conv2D(32,(3,3),padding='valid'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))
```

```
model.add(Conv2D(64,(3,3),padding='valid'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Conv2D(64,(3,3),padding='valid'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2,2)))
```

```
model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(Dense(classer))
model.add(Activation('softmax'))
```

```
#du lieu sau khi qua cac conv
model.summary()
```

```
#tang cuong du lieu
aug = ImageDataGenerator(rotation_range=0.18,zoom_range=0.15,
                         width_shift_range=0.2,height_shift_range=0.2,horizontal_flip=True)
```

```
learning_rate = 0.01
epochs = 10
```

```
otp = SGD(learning_rate=learning_rate,momentum=0.9)
```

```
model.compile(optimizer=otp,loss="categorical_crossentropy",metrics=["accuracy"])
```

```
batch_size = 400
```

```
H = model.fit_generator(aug.flow(trainX, trainY, batch_size=batch_size),validation_data=(testX, testY),
                        steps_per_epoch=trainX.shape[0]//batch_size, epochs=epochs, verbose=1)
```
